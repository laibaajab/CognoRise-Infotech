```python
import warnings
warnings.filterwarnings('ignore')


import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(style="white", color_codes = True)
```

```python
iris =pd.read_csv("IRIS.csv")
iris.head()
```

```python
iris["species"].value_counts()
```

```python
sns.FacetGrid(iris, hue="species",height=6).map(plt.scatter , "petal_length" , "sepal_width").add_legend()
```

# Logistic Regression

# converting categorical variables into numbers

```python
flower_mapping = {"setosa" :0, "versicolor" :1, "virginica":2}
iris["species"]= iris ["species"].map(flower_mapping)
```

```python
iris.head()
```

# Preparing inputs and outputs

```python
X=iris[['sepal_length' , 'sepal_width', 'petal_length', 'petal_width']].values
y=iris[["species"]].values
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X,y)
```

# Accuracy

```python
model.score(X,y)
```

# make predictions

```python
expected = y
predicted = model.predict(X)
predicted
```

# summarize the fit of the model

```python
from sklearn import metrics
```

```python
print(metrics.classification_report(predicted,expected))
```

```python
print(metrics.confusion_matrix(predicted,expected))
```
